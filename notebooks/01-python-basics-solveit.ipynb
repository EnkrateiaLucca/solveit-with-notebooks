# 🧩 Python Basics — the SolveIt way
### A guided tour of our notebook-native learning environment

Welcome! This is the same Python-basics class you know, rebuilt as a
**SolveIt-style** environment that lives entirely inside Jupyter.

SolveIt (by Jeremy Howard) is built on George Pólya's *How to Solve It* (1945).
The whole philosophy fits in four lines:

1. **Dialogue, not prompting** — talk *with* the AI over many small turns.
2. **Pólya's loop** — Understand → Plan → Do (tiny steps) → Review.
3. **Tiny steps + feedback** — write 1–2 lines, run it, look, repeat.
4. **The human is the agent** — the AI *proposes*; **you** read, edit, and run.

This notebook is also your **manual**: it teaches the tool while you use it.

> ## 🗺️ How to use this notebook
>
> Every cell is labelled so you always know what to do with it:
>
> | Badge | Meaning |
> |---|---|
> | ▶️ **Run** | Just run it and read the output. |
> | 🔎 **Notice** | Pause — look at *why* the output is what it is. |
> | 💬 **Ask** | An AI cell. Run it, read the reply, then ask a follow-up of your own. |
> | ✍️ **Your turn** | You write/edit the code before running. The whole point. |
>
> **Golden rule:** never just run the AI cells passively. After each one, type
> your *own* follow-up question in a new cell. The learning is in the dialogue.
>
> Go top to bottom. Don't rush. Keep cells small.

## 4 · A real problem, Pólya-style — `solve()`  🧩

This is the heart of SolveIt. Instead of being handed the *Time Travel
Calculator* solution, you'll **build it yourself in tiny steps**, with a coach.

▶️ **Run** to start the session:

````python
p = solve("Ask the user their age and how many years to travel forward, "
          "then print their future age.")
````

**Stage 1 — Understand.** 💬 Run it. Read how it restates the problem and
names the inputs/outputs. *This step matters most — don't skip it.*

````python
p.understand()
````

**Stage 2 — Plan.** 💬 A plain-English plan, each step tiny enough for 1–2 lines:

````python
p.plan()
````

**Stage 3 — Do.** 💬 It gives only the *smallest next step*...

````python
p.step()
````

✍️ **Your turn** — write that step *yourself* and run it. Look at the result:

````python
# 👇 your tiny step goes here
age = int(input("Your age: "))
````

💬 **Ask for the next step**, telling the coach where you are. Repeat this
run-look-step rhythm until it works:

````python
p.step("I have age as an int, now I need the years to travel")
````

**Stage 4 — Review.** 💬 Once you have a working version, *look back*. Paste
your final code in and let it critique:

````python
p.review("""age = int(input("Your age: "))
years = int(input("Years forward: "))
print(f"In {years} years you will be {age + years}")""")
````

## 5 · The human is the agent — `propose()`  ✍️

`propose()` asks the AI for code and drops it into a **new cell below** — it
**does not run it**. You read it, edit it, and decide. This keeps *you* in
control instead of letting the AI run wild.

▶️ **Run** it, then look at the new cell that appears underneath:

````python
propose("count how many words are in the sentence stored in a variable called text")
````

````python
# 🤖 PROPOSED (review & edit before running)
text = "Hello world this is Python"
word_count = len(text.split())  # split() breaks text into a list of words
print(word_count)
````

🔎 **Notice** a new code cell appeared below with proposed code (commented as
`🤖 PROPOSED`). ✍️ **Your turn:** read it, define a `text` variable, edit if
needed, *then* run it. Never run proposed code blindly.

## 9 · Let the AI edit your code — `%%edit`  ✂️

Write code in a cell, then ask the AI to change it **in place**, with a comment
explaining what it did. You stay in control: it rewrites the cell, you review
and re-run.

▶️ **Run this cell** (it has a clumsy loop). The `%%edit` line says what you want:

````python
# ✏️ EDIT: Replaced range(len()) loop with direct iteration over list and added a docstring explaining what the function does
def totals(xs):
    """Return a new list with each number from xs doubled."""
    result = []
    for x in xs:
        result.append(x * 2)
    return result
````

🔎 **Notice** the cell's contents got **replaced** with the improved code,
topped by a `# ✏️ EDIT:` comment explaining the change. Re-run it to use it.

Prefer the edit in a *new* cell instead of in place? Use the function form:

````python
edit('"hello" + 100', "fix this so it concatenates correctly")
````

````python
# ✏️ EDIT: Converted the number 100 to a string so it can be concatenated with 'hello'
"hello" + str(100)
````
